```python
# Create new columns to account for the age of a house, how many years since the remodeling, 
# overall quality and condition (combined).
# Dropping year columns

df_train['House_Age'] = df_train['Yr_Sold'] - df_train['Year_Built']
df_test['House_Age'] = df_test['Yr_Sold'] - df_test['Year_Built']
df_train.loc[1885, 'House_Age'] = 0

df_train['Remod/Add_Age'] = df_train['Yr_Sold'] - df_train['Year_Remod/Add']
df_test['Remod/Add_Age'] = df_test['Yr_Sold'] - df_test['Year_Remod/Add']
df_train.loc[1719, 'Remod/Add_Age'] = 0
df_train.loc[1885, 'Remod/Add_Age'] = 0


df_train['Qual_Cond'] = (df_train['Overall_Qual'] * df_train['Overall_Cond'])
df_test['Qual_Cond'] = (df_test['Overall_Qual'] * df_test['Overall_Cond'])
    
    
df_train.drop(columns=['Year_Built', 'Year_Remod/Add', 'Garage_Yr_Blt'], inplace=True)
df_test.drop(columns=['Year_Built', 'Year_Remod/Add', 'Garage_Yr_Blt'], inplace=True)    
```

```python
# Converting the following columns from numerical into categorical 

cat_cols = ['Bedroom_AbvGr', 'Bsmt_Full_Bath', 'Bsmt_Half_Bath', 'Fireplaces', 'Full_Bath', 'Garage_Cars', \
            'Half_Bath', 'Id', 'Kitchen_AbvGr', 'Mo_Sold', 'MS_SubClass', 'TotRms_AbvGrd', 'Yr_Sold'] 

for x in cat_cols:
    df_train[x] = df_train[x].astype(str)
    df_test[x] = df_test[x].astype(str)
```

```python
# Replacing NaN in numerical columns.
# A function is developed that it assumes that a linear relationship exists between the columns with  
# missing values and the Gr_Liv_Area for that specific row. 


num_cols = sorted(df_train.loc[:, df_train.dtypes != object].columns)
features = num_cols.copy()
index = features.index('SalePrice')
features.pop(index)



for item in features:
    if item == 'Garage_Age':
        if df_train[item].isnull().sum()!=0:
            index_train = df_train[item].index[df_train[item].apply(np.isnan)].tolist()
            for x in index_train:
                df_train.loc[x, item] = df_train.loc[x, 'House_Age']
            
        if df_test[item].isnull().sum()!=0:   
            index_test = df_test[item].index[df_test[item].apply(np.isnan)].tolist()
            for x in index_test:
                df_test.loc[x, item] = df_test.loc[x, 'House_Age']
    elif item == 'SalePrice':
        pass
    else:
        if df_train[item].isnull().sum()!=0:
            index_train = df_train[item].index[df_train[item].apply(np.isnan)].tolist()

            intercept, coeff = lr_fix(df_train, item, 'Gr_Liv_Area')
            for x in index_train:
                df_train.loc[x, item] = intercept + coeff *  df_train.loc[x, 'Gr_Liv_Area']

        if df_test[item].isnull().sum()!=0:
            index_test = df_test[item].index[df_test[item].apply(np.isnan)].tolist()

            intercept, coeff = lr_fix(df_train, item, 'Gr_Liv_Area')
            for x in index_test:
                df_test.loc[x, item] = intercept + coeff *  df_test.loc[x, 'Gr_Liv_Area']  
```

## Scaling and Normalizing Numerical Columns

```python
# Create a list of numerical columns that have non zero values. Log transformation will be applied
# on these columns. For the rest of the numerical columns, standard scalar will be applied.
# Log Transformation will be applied on 'SalePrice' column in Training set

log_list = []
for item in sorted(numerical_features):
    if (df_train[item].min()>0) & (df_test[item].min()>0) & (item!='Overall_Qual') & (item!='Overall_Cond') & \
    (item!='Qual_Cond') & (item!='House_Age') & (item!='Remod/Add_Age'):
        log_list.append(item) 
        df_train[item] = np.log(df_train[item])
        df_test[item] = np.log(df_test[item])    
    else:
        if item not in ['House_Age', 'Remod/Add_Age', 'Overall_Qual', 'Overall_Cond', 'Qual_Cond']:
            log_list.append(item)
            df_train[item] = df_train[item] + 1
            df_test[item] = df_test[item] + 1

            df_train[item] = np.log(df_train[item])
            df_test[item] = np.log(df_test[item])    
        else:
            if item not in ['Overall_Qual', 'Overall_Cond', 'Qual_Cond']:
                standard_scaler = []
                standard_scaler.append(item)
                mean = np.mean(df_train[item])
                std = np.std(df_train[item])
                df_train[item] = (df_train[item]-mean)/std
                df_test[item] = (df_test[item]-mean)/std

                
# log-transforming the SalePrice column in Training
df_train['SalePrice'] = np.log(df_train['SalePrice'])                
```

```python
# Training the Lasso, Ridge, and linear regression models


model_features = []
model_features = numerical_features + categorical_features


signif_features = model_features.copy()


for i in range(1):
    print('Feature size')
    print(len(signif_features), len(model_features))
    
    X = df_train[signif_features]
    y = df_train['SalePrice']    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


    l_alphas = np.logspace(-5, -3, 50)
    #l_alphas = np.linspace(1e-4, 0, 50)
    lasso_cv = LassoCV(alphas=l_alphas, cv=5, max_iter=100000)
    lasso_cv.fit(X_train, y_train)
    print('Lasso')
    print(round(lasso_cv.score(X_train, y_train),4), \
    round(cross_val_score(lasso_cv, X_train, y_train, cv = 5).mean(),4), \
    round(lasso_cv.score(X_test, y_test), 4), \
    round(np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(lasso_cv.predict(X_test)))),4), lasso_cv.alpha_)

    
    
    #r_alphas = np.logspace(0, 5, 50)
    r_alphas = np.linspace(0, 5, 50)
    ridge_cv = RidgeCV(alphas= r_alphas, cv=5, scoring = 'neg_mean_squared_error')
    ridge_cv.fit(X_train, y_train)
    print('Ridge')
    print(round(ridge_cv.score(X_train, y_train),4), \
    round(cross_val_score(ridge_cv, X_train, y_train, cv = 5).mean(),4), \
    round(ridge_cv.score(X_test, y_test), 4), \
    round(np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(ridge_cv.predict(X_test)))),4), ridge_cv.alpha_ ) 



    lr = LinearRegression()
    lr.fit(X_train, y_train)
    print('LR')
    print(round(lr.score(X_train, y_train),4), \
    round(cross_val_score(lr, X_train, y_train, cv = 5).mean(),4), \
    round(lr.score(X_test, y_test), 4), \
    round(np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(lr.predict(X_test)))),4))

```
